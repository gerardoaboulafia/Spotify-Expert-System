```python
import requests
import pandas as pd
import time
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

cid = ''
secret = ''

client_credentials_manager = SpotifyClientCredentials(client_id=cid, client_secret=secret)
sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager)
```

```python
# Playlist 1
playlist_url = 'https://open.spotify.com/playlist/5yrd7jrEcmrBEwyOYwmc6u?si=d8ba2b21426d49c2'
playlist_id = playlist_url.split('/')[-1].split('?')[0]  # Extraer el ID de la playlist
```

```python
# Extraer canciones de la playlist
def get_playlist_tracks(playlist_id, max_tracks=2000):
    offset = 0
    limit = 100  # Máximo permitido por la API
    tracks = []
    
    while offset < max_tracks:
        results = sp.playlist_tracks(playlist_id, limit=limit, offset=offset)
        items = results['items']
        tracks.extend(items)
        offset += limit
        
        # Detener si ya no hay más canciones
        if len(items) < limit:
            break
    
    return tracks

# Extraer información de canciones, incluyendo géneros, artistas y características adicionales
def process_tracks(tracks):
    songs_data = []
    
    for track_item in tracks:
        track = track_item['track']
        if track:  # Asegurarse de que no haya datos nulos
            # URI de la canción y el artista
            track_uri = track['uri']
            artist_uri = track['artists'][0]['uri']
            
            # Información de la canción
            track_name = track['name']
            album_name = track['album']['name']
            track_popularity = track['popularity']
            
            # Información del artista
            artist_name = track['artists'][0]['name']
            artist_info = sp.artist(artist_uri)
            artist_popularity = artist_info.get('popularity', 0)
            artist_genres = artist_info.get('genres', ["unknown"])
            
            # Añadir la información procesada al dataset
            songs_data.append({
                'name': track_name,
                'artist': artist_name,
                'genre': artist_genres if artist_genres else "unknown",
                'album': album_name,
                'popularity': track_popularity,
                'artist_popularity': artist_popularity,
                'track_uri': track_uri,
                'artist_uri': artist_uri
            })
    
    return songs_data

# Obtener características de audio para varias canciones
def get_audio_features(track_ids):
    features = []
    for i in range(0, len(track_ids), 100):  # Procesar en bloques de 100
        batch = track_ids[i:i+100]
        features.extend(sp.audio_features(batch))
    return features

# Extraer canciones de la playlist
print("Extrayendo canciones de la playlist...")
tracks = get_playlist_tracks(playlist_id, max_tracks=2000)

# Procesar canciones y obtener géneros
print("Procesando canciones y obteniendo géneros...")
songs_data = process_tracks(tracks)

# Obtener características de audio para las canciones
print("Obteniendo características de audio...")
track_ids = [song['track_uri'] for song in songs_data]
audio_features = get_audio_features(track_ids)

# Combinar características de audio con datos de las canciones
for i, feature in enumerate(audio_features):
    if feature:  # Asegurarse de que no haya datos nulos
        songs_data[i].update({
            'danceability': feature['danceability'],
            'energy': feature['energy'],
            'key': feature['key'],
            'loudness': feature['loudness'],
            'mode': feature['mode'],
            'speechiness': feature['speechiness'],
            'acousticness': feature['acousticness'],
            'instrumentalness': feature['instrumentalness'],
            'liveness': feature['liveness'],
            'valence': feature['valence'],
            'tempo': feature['tempo']
        })

# Crear un DataFrame para una vista tabular de los datos
df = pd.DataFrame(songs_data)
```

```python
import ast

def normalize_genre_enhanced(genres):
    # Asegúrate de que la entrada sea un string evaluable
    try:
        genres_list = ast.literal_eval(genres) if isinstance(genres, str) else genres
        genres_str = ', '.join(genres_list) if isinstance(genres_list, list) else str(genres_list)
    except (ValueError, SyntaxError):
        genres_str = str(genres)

    # Reglas de normalización mejoradas
    if 'pop' in genres_str or 'high vibe' in genres_str or 'gen z singer-songwriter' in genres_str:
        return 'pop'
    elif 'rock' in genres_str or 'metal' in genres_str or 'punk' in genres_str:
        return 'rock'
    elif 'indie' in genres_str or 'deep new americana' in genres_str:
        return 'indie'
    elif 'reggaeton' in genres_str or 'latino' in genres_str or 'urbano' in genres_str or 'moombahton' in genres_str or \
         'techengue' in genres_str or 'dembow' in genres_str or 'cubaton' in genres_str:
        return 'reggaeton'
    elif 'r&b' in genres_str:
        return 'r&b'
    elif 'hip hop' in genres_str or 'rap' in genres_str or 'drill' in genres_str or 'beats' in genres_str or \
         'boom bap' in genres_str:
        return 'rap'
    elif 'soul' in genres_str or 'motown' in genres_str or 'afrofuturism' in genres_str:
        return 'soul'
    elif 'alternative' in genres_str or 'alt' in genres_str or 'movie tunes' in genres_str:
        return 'alternative'
    elif 'electro' in genres_str or 'dance' in genres_str or 'edm' in genres_str or 'electronic' in genres_str or \
         'house' in genres_str:
        return 'electro'
    elif 'salsa' in genres_str or 'mambo' in genres_str or 'boogaloo' in genres_str or 'tropical' in genres_str or \
         'bachata' in genres_str or 'merengue' in genres_str or 'orquesta tropical' in genres_str:
        return 'salsa'
    elif 'folk' in genres_str or 'cantautor' in genres_str or 'nouvelle chanson francaise' in genres_str or \
         'chanson' in genres_str or 'musica chihuahuense' in genres_str or 'sad sierreno' in genres_str or \
         'musica portuguesa contemporanea' in genres_str:
        return 'folk/latino'
    elif 'jazz' in genres_str:
        return 'jazz'
    elif genres_str.lower() == 'unknown':
        return 'unknown'
    else:
        return genres_str

# Aplicar la función actualizada de normalización al nuevo dataset
data['normalized_genre_enhanced'] = data['genre'].apply(normalize_genre_enhanced)

# Verificar si quedan géneros sin clasificar adecuadamente
remaining_genres_enhanced = data[data['normalized_genre_enhanced'].apply(lambda x: ',' in x or x not in {
    'pop', 'rock', 'indie', 'reggaetón/latino', 'r&b', 'hip hop/rap',
    'soul', 'alternative', 'electro', 'folk/latino', 'jazz', 'salsa', 'unknown'})]['normalized_genre_enhanced'].unique()
```

```python
# Eliminamos el género unknown, y los géneros que tengan menos de 20 canciones.
data = data[data['normalized_genre_enhanced'] != 'unknown']
data = data.groupby('normalized_genre_enhanced').filter(lambda x: len(x) >= 20)
```

### Crear base de conocimiento en prolog

```python
from pyswip import Prolog

# Inicializa Prolog
prolog = Prolog()

# Función para regenerar hechos Prolog desde el DataFrame
def dataframe_to_prolog_facts_safe(df, file_path):
    with open(file_path, 'w') as f:
        for _, row in df.iterrows():
            # Escapar comillas simples en los valores
            row = row.apply(lambda x: str(x).replace("'", "\\'") if isinstance(x, str) else x)
            fact = (
                f"song(genre({row['genre']}), name('{row['name']}'), artist('{row['artist']}'), "
                f"album('{row['album']}'), popularity({row['popularity']}), "
                f"artist_popularity({row['artist_popularity']}), "
                f"danceability({row['danceability']}), energy({row['energy']}), "
                f"key({row['key']}), loudness({row['loudness']}), mode({row['mode']}), "
                f"speechiness({row['speechiness']}), acousticness({row['acousticness']}), "
                f"instrumentalness({row['instrumentalness']}), liveness({row['liveness']}), "
                f"valence({row['valence']}), tempo({row['tempo']})).\n"
            )
            f.write(fact)

# Regenerar el archivo .pl
output_file_path_safe = "/Users/gerardoaboulafia/Library/Mobile Documents/com~apple~CloudDocs/UCA/Documentos/Cuatrimestre 4/Algoritmia/Sistema Experto/music_knowledge_base_safe.pl"
dataframe_to_prolog_facts_safe(data_cleaned, output_file_path_safe)
```

### Testing del modelo

Para testear el funcionamiento del modelo, vamos a intentar obtener 10 canciones similares a través de KNN. 

```python
from sklearn.neighbors import NearestNeighbors

# Filtrar solo las características numéricas relevantes para calcular similitud
features = [
    "Popularity", "Danceability", "Energy", "Key", "Loudness", 
    "Mode", "Speechiness", "Acousticness", "Instrumentalness", "Liveness", "Valence", "Tempo"
]
data_k = df[features]
```

```python
# Normalizar los datos para evitar que los rangos influyan demasiado
data_normalized = (data_k - data_k.min()) / (data_k.max() - data_k.min())
```

```python
# Configurar el modelo k-NN
knn = NearestNeighbors(n_neighbors=10, metric="euclidean")
knn.fit(data_normalized)
```

```python
# Elegir una canción por su índice o nombre
song_name = "Come to Life"
nombre_artista = "Kanye West"  # Reemplaza con el nombre del artista que estás buscando

# Filtrar por el nombre de la canción y el artista, sin sensibilidad a mayúsculas
filtered_df = df[(df['Name'].str.lower() == song_name.lower()) & 
                 (df['Artist'].str.lower() == nombre_artista.lower())]

# Ordenar por popularidad en orden descendente
filtered_df = filtered_df.sort_values(by='Popularity', ascending=False)

# Obtener el índice de la canción más popular que cumpla con los criterios
if not filtered_df.empty:
    chosen_song_index = filtered_df.index[0]
else:
    chosen_song_index = None  # Si no hay coincidencias
```

```python
# Encontrar las 10 canciones más similares
distances, indices = knn.kneighbors([data_normalized.iloc[chosen_song_index]])
```

```python
# Obtener los resultados
similar_songs = df.iloc[indices[0]]
similar_songs["Distance"] = distances[0]

# Mostrar las canciones similares
(similar_songs)
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd

cid = 'c48a716678f5441e926cbc0fe5a046b0'
secret = 'b3d86e728f6d46649cf7154a8201f55b'

client_credentials_manager = SpotifyClientCredentials(client_id=cid, client_secret=secret)
sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager)
```

```python
def buscar_canciones(nombre_cancion):
    try:
        # Buscar las primeras 10 coincidencias
        resultados = sp.search(q=nombre_cancion, type='track', limit=10)
        canciones = []
        
        for item in resultados['tracks']['items']:
            # Obtener los datos básicos
            nombre = item['name']
            artista = item['artists'][0]['name']
            album = item['album']['name']
            popularidad = item['popularity']
            
            # Obtener análisis de audio para la canción
            audio_features = sp.audio_features(item['id'])[0]
            if audio_features:  # Verificar que existan análisis de audio
                canciones.append({
                    'Name': nombre,
                    'Artist': artista,
                    'Album': album,
                    'Popularity': popularidad,
                    'ArtistPopularity': sp.artist(item['artists'][0]['id'])['popularity'],
                    'Danceability': audio_features['danceability'],
                    'Energy': audio_features['energy'],
                    'Key': audio_features['key'],
                    'Loudness': audio_features['loudness'],
                    'Mode': audio_features['mode'],
                    'Speechiness': audio_features['speechiness'],
                    'Acousticness': audio_features['acousticness'],
                    'Instrumentalness': audio_features['instrumentalness'],
                    'Liveness': audio_features['liveness'],
                    'Valence': audio_features['valence'],
                    'Tempo': audio_features['tempo']
                })
        
        # Crear un DataFrame con los resultados
        df = pd.DataFrame(canciones)
        return df
    except Exception as e:
        print(f"Error al buscar canciones: {e}")
        return None
```
